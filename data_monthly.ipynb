# 对scaled_nav做分类

```python
data2 = pd.read_csv('./dataset/scaled_nav.csv')
data2.drop('Unnamed: 0',axis=1,inplace=True)
data2
```

```python

sample_data=data2.T
sample_data.columns = sample_data.loc['fund_id']
sample_data.drop(['fund_id'],inplace=True)
sample_data.plot(legend=False,figsize=(40,30))
```

```python
ls=['HF000025PG','HF00001ZZI','HF000012AM'
    ,'HF00001AUE','HF000024Y9','HF00004NLB']
tmp=data2[data2.fund_id.isin(ls)]
tmp = tmp.T
tmp.columns = tmp.loc['fund_id']
tmp.drop(['fund_id'],inplace=True)
tmp.plot(legend=False,figsize=(40,30))
```

```python
from sklearn.cluster import DBSCAN
```

```python
X = data2.drop(['fund_id','month1'],axis=1)
y_pred = DBSCAN(eps=0.1,min_samples=4).fit_predict(X)
```

```python
dic={}
for i in list(y_pred):
    dic.setdefault(i,0)
    dic[i]+=1
dic
```

```python
data2['label'] = pd.Series(y_pred)
```

```python
data2.to_csv('./dataset/dbscan.csv')
```

```python
tmp = data2[data2['label']==6]
tmp = tmp.T
tmp.columns = tmp.loc['fund_id']
tmp.drop(['fund_id','label'],inplace=True)
tmp.plot(legend=False,figsize=(40,30),c='r')
```

```python
plt.show()
```

```python
data2
```

```python
rtn_list = ['rtn2','rtn3','rtn4','rtn5'
            ,'rtn6','rtn7','rtn8','rtn9',
           'rtn10','rtn11','rtn12']
```

```python
month_list = list(data2.columns)[1:-1]
```

```python
for i in range(11):
    data2[rtn_list[i]] = (data2[month_list[i+1]]-data2[month_list[i]])/data2[month_list[i]]        
```

```python
data2.head()
```

```python
tmp=1.
for rtn in rtn_list:
    tmp*=(1+data2[rtn])
data2['rtn_year'] = tmp**(12/11)-1
```

```python
data2.head()
```

```python
#最大回撤
max_drawdown_ls = []
for index in range(len(data2)):  
    tmp_series = data2.iloc[index]
    max_drawdown = 0
    for i in range(len(month_list)-1):
        for j in range(i+1,len(month_list)):
            tmp = (tmp_series[month_list[i]]-tmp_series[month_list[j]])/tmp_series[month_list[i]]
            if tmp>=max_drawdown:
                max_drawdown = tmp
                pre = i
                tail = j
    max_drawdown_ls.append(max_drawdown)
data2['max_drawdown'] = pd.Series(max_drawdown_ls)
```

```python
# 计算波动率

np.std([1,1,1])
```

```python
#计算波动率
volatility_ls = []
for index in range(len(data2)):  
    tmp_series = data2.iloc[index]
    ls = []
    for rtn in rtn_list:
        ls.append(tmp_series[rtn])
    volatility_ls.append(np.std(ls)*np.sqrt(12))
data2['volatility'] = pd.Series(volatility_ls)
```

```python
data2.head()
```

```python
# 计算夏普
data2['sharpe'] = (data2['rtn_year'] - 0.4)/data2['volatility']
```

```python
data2.head()
```

```python
# 通过类别标签输出dataframe(columns = [labe,rtn_year,max_drawdown,volatility,sharpe])

for i in list(range(min(data2.label),max(data2.label)+1)):
    label_data = data2[data2['label']==i][['label','rtn_year','max_drawdown','volatility','sharpe']]
    label_data.to_csv(('./DBSCAN/'+'DBSCAN'+str(i)+'.csv'))
```
