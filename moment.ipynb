```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('../dataset/data_selected_monthly_3year.xlsx')
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
rtn_list =[]
for i in range(1,36):
    rtn_list.append('rtn'+str(i))
```

```python
month_list = list(data.columns[1:])
```

```python
month_list[1]
```

```python
for i in range(1,len(month_list)):
    data[month_list[i]] = data[month_list[i]]/data['month1']
```

```python
data.head()
```

```python
moment_list = ['moment1','moment2','moment3','moment4']
```

```python
data['moment1'] = data[rtn_list].mean(axis=1)
```

```python
for month in month_list:
    data.drop(month,axis=1,inplace = True)
```

```python
data.head()
```

```python
for i in range(2,11):
    tmp_sum = 0
    for rtn in rtn_list:
        tmp_sum+=((data[rtn] - data['moment1'])**i)
    data['moment'+str(i)] = tmp_sum/35
```

```python
for rtn in rtn_list:
    data.drop(rtn,axis=1,inplace = True)
```

```python
data.to_csv('./dataset/moment10.csv')
```

# 对1-10阶矩数据进行聚类

```python
# 标准化
moment_list = list(data.columns)[1:]
for moment in moment_list:
    data[moment] = (data[moment] - data[moment].mean())/data[moment].std()
```

```python
data.moment6.describe()
```

```python
for i in range(7,11):
    data.drop('moment'+str(i),axis=1,inplace=True)
```

```python
data
```

```python
data2 = data.copy(deep=True)
```

```python
data2.drop('month1',axis=1,inplace=True)
```

```python
#spectral
from sklearn.cluster import SpectralClustering
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import Birch
from sklearn.cluster import MeanShift
from sklearn.cluster import OPTICS
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN
X = data2.drop(['fund_id'],axis=1)
y_pred = KMeans(n_clusters=20).fit_predict(X)
data2['Kmeans_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = SpectralClustering(n_clusters=20,affinity='sigmoid').fit_predict(X)
data2['Spectral_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = DBSCAN(eps=0.05,min_samples=3).fit_predict(X)
data2['DBSCAN_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = AgglomerativeClustering(n_clusters=20,linkage='ward').fit_predict(X)
data2['Hierarchical_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = Birch(n_clusters=20).fit_predict(X)
data2['Birch_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = MeanShift().fit_predict(X)
data2['MeanShift_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = OPTICS().fit_predict(X)
data2['OPTICS_nav'] = pd.Series(y_pred)

X = data2.drop(['fund_id'],axis=1)
y_pred = GaussianMixture(n_components=20).fit_predict(X)
data2['GaussianMixture_nav'] = pd.Series(y_pred)
```

```python
for i in range(1,len(month_list)):
    data2.drop(month_list[i],inplace=True,axis=1)
```

```python
data2.drop('fund_id',axis=1,inplace=True)
```

```python
data2.to_csv('./dataset/clusters_base_on_moment.csv')
```
