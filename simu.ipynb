```python
#构造fund_id、起始时间、结束时间的字典，
#键为fund_id,值为start与end组成的列表
time_span_dict = {}
time_span_dataframe = pd.DataFrame(columns=['fund_id','start','end'])
for fund_id in fund_id_list:
    tmp_data_frame = useful_data[useful_data['fund_id']==fund_id]
    end = list(tmp_data_frame['fund_day'])[-1]
    start = list(tmp_data_frame['fund_day'])[0]
    value = [start,end]
    time_span_dict[fund_id]=value

for i,v in time_span_dict.items():
    time_span_dataframe= time_span_dataframe.append(
        [{'fund_id':i,'start':v[0],'end':v[1]}],ignore_index=True)
time_span_dataframe['start'] = pd.to_datetime(time_span_dataframe['start'])
time_span_dataframe['end'] = pd.to_datetime(time_span_dataframe['end'])
time_span_dataframe

```

```python
# 暴力法筛选出开始日期和结束日期，使得区间中的数据量最大
max_tmp = 0

for i in start_list:
    for j in end_list:
        count = 0
        for fund_id in fund_id_list:
            tmp_data_frame = useful_data[useful_data['fund_id']==fund_id]
            left = tmp_data_frame.iloc[0]['fund_day']
            right = tmp_data_frame.iloc[len(tmp_data_frame)-1]['fund_day']
            if left <=i and right>=j:
                count+=1
        if count>=max_tmp:
            max_tmp = count
            start0 = i
            end0 = j
print('开始时间为:',start0)
print('结束时间为:',end0)
print('区间内的fund个数:',max_tmp)
```
